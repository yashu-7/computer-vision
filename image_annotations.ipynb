```python
import cv2
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
lake_img_path = 'images\\nice_view_lake.jpg'
```

```python
lake_img = cv2.imread(lake_img_path)
Image(lake_img_path)
```

```python
plt.imshow(lake_img[:,:,::-1])
```

```python
lake_img_copy = lake_img.copy()
cv2.line(lake_img_copy, (0,0), (3000,4000), (255,0,255), 15, lineType=cv2.LINE_AA)
plt.imshow(lake_img_copy[:,:,::-1])
```

```python
print(lake_img_copy.shape)
h,w,c = lake_img_copy.shape
cv2.circle(lake_img_copy, (w//2,h//2), 100, (0,255,255), 15)
plt.imshow(lake_img_copy[:,:,::-1])
```

```python
cv2.rectangle(lake_img_copy, (w//2-50,h//2-50), (w//2+50,h//2+50), (255,255,0), 15)
plt.imshow(lake_img_copy[:,:,::-1])
```

```python
text = "Nice Lake View"
fontface = cv2.FONT_HERSHEY_PLAIN
# Suported lineTypes = LINE_AA,LINE_4, LINE_8
cv2.putText(lake_img_copy, text, (1000,700), fontface, 12.5, (85,170,255), 25, lineType=cv2.LINE_AA)
cv2.putText(lake_img_copy, text, (1000,700), fontface, -12.5, (85,170,255), 25, lineType=cv2.LINE_AA)
plt.imshow(lake_img_copy[:,:,::-1])
```
