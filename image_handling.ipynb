```python
lake_img_path = image_folder_path+'\lake.jpg'
```

```python
lake_img = cv2.imread(lake_img_path,1)
```

```python
plt.imshow(lake_img)
```

```python
lake_img_reversed = lake_img[:, :, ::-1]
```

```python
plt.imshow(lake_img_reversed)
```

```python
b,g,r = cv2.split(lake_img)
lake_img = cv2.merge((b,g,r))
```

```python
plt.figure(figsize=(20, 4))
plt.subplot(1, 4, 1); plt.imshow(b)
plt.subplot(1, 4, 2); plt.imshow(g)
plt.subplot(1, 4, 3); plt.imshow(r)
plt.subplot(1, 4, 4); plt.imshow(lake_img_reversed)
```

```python
lake_img_rgb = cv2.cvtColor(lake_img, cv2.COLOR_BGR2RGB)
plt.imshow(lake_img_rgb)
```

```python
r, g, b = cv2.split(lake_img_rgb)
```

```python
r+=50
```

```python
new_lake_img_rgb = cv2.merge((r,g,b))
plt.imshow(new_lake_img_rgb)
```
