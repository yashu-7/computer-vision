```python
import cv2
import matplotlib.pyplot as plt
```

```python
source = 'examples/turtle_in_sea.mp4'
```

```python
cap = cv2.VideoCapture(source)
```

```python
if cap.isOpened() == False:
    print("ERROR OPENING FILE")
```

```python
ret, frame = cap.read()
```

```python
plt.imshow(frame[:,:,::-1])
```

```python
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
```

```python
output_avi = cv2.VideoWriter('saved_examples\\turtle_in_sea.avi',cv2.VideoWriter_fourcc('M','J','P','G'),10,(frame_width,frame_height))
output_mp4 = cv2.VideoWriter('saved_examples\\turtle_in_sea.mp4',cv2.VideoWriter_fourcc(*"XVID"),10,(frame_width,frame_height))
```

```python
while cap.isOpened():
    ret,frame = cap.read()
    if ret:
        output_avi.write(frame)
        output_mp4.write(frame)
    else:
        break
```

```python
cap.release()
output_avi.release()
output_mp4.release()
```
