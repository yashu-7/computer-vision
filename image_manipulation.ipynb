```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
lake_img_path = 'images/lake.jpg'
```

```python
lake_img = cv2.imread(lake_img_path)
lake_img = lake_img[:,:,::-1]
plt.imshow(lake_img)
```

```python
lake_img_pixel = lake_img[0,0]
print(lake_img_pixel)
```

```python
lake_img_copy = lake_img.copy()
lake_img_copy[0:180:2,0:286:2] = [255,255,255]
plt.imshow(lake_img_copy) 
```

```python
cropped_lake_img = lake_img[10:80, 90:160]
plt.imshow(cropped_lake_img)
```

```python
scaled_cropped_lake_img = cv2.resize(cropped_lake_img,None,fx=1.5,fy=1.5)
plt.imshow(scaled_cropped_lake_img)
```

```python
desired_width = 150
desired_height = 150
dim = (desired_width,desired_height)
resized_lake_image = cv2.resize(cropped_lake_img, dim,interpolation=cv2.INTER_AREA)
plt.imshow(resized_lake_image)
```

```python
scaled_cropped_lake_img_horz = cv2.flip(scaled_cropped_lake_img,1)
scaled_cropped_lake_img_vert = cv2.flip(scaled_cropped_lake_img,0)
scaled_cropped_lake_img_both = cv2.flip(scaled_cropped_lake_img,-1)

plt.figure(figsize=(20,5))
plt.subplot(1,4,1);plt.imshow(scaled_cropped_lake_img);plt.title('Orignal')
plt.subplot(1,4,2);plt.imshow(scaled_cropped_lake_img_horz);plt.title('Horizontal Flip')
plt.subplot(1,4,3);plt.imshow(scaled_cropped_lake_img_vert);plt.title('Vertical Flip')
plt.subplot(1,4,4);plt.imshow(scaled_cropped_lake_img_both);plt.title('Both Flip')
```
